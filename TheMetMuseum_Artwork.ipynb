```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#to print the specific main image
image_tag = soup.find('section', {'id': 'artwork-section'}).find('img')
if image_tag:
    image_url = image_tag['src']
else:
    image_url = "Image not found"
```

```python
print("Image URL:", image_url)
```

```python
#to print the specific description text
description_tag = soup.find('div', {'class': 'artwork__intro__desc js-artwork__intro__desc'})
if description_tag:
    # Get the description text inside the <p> tag
    description = description_tag.get_text(strip=True)
else:
    description = "Description not found"
```

```python
print("Description:", description)
```

```python
#to print the overview section from the details 
overview_tag = soup.find('section', {'id': 'overview'})
if overview_tag:
    # Find all <p> tags with the class "artwork-tombstone--item"
    overview_items = overview_tag.find_all('p', {'class': 'artwork-tombstone--item'})
    overview = "\n".join([f"{item.find('span', {'class': 'artwork-tombstone--label'}).get_text(strip=True)}: {item.find('span', {'class': 'artwork-tombstone--value'}).get_text(strip=True)}" for item in overview_items])
else:
    overview = "Overview not found"
```

```python
print("Overview:", overview)
```

```python
# Writing data to CSV
header = ['Image URL', 'Description', 'Overview']
data = [image_url, description, overview]
```

```python
# Write to CSV
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # Write the header
    writer.writerow(data)    # Write the extracted data
```

```python
def scrape_artwork_info(url):
    # Fetch the webpage
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')


# Extract overview details
    overview_section = soup.find('section', {'id': 'overview'})
    overview_details = {
        'Title': "Not available",
        'Artist': "Not available",
        'Date': "Not available",
        'Medium': "Not available",
        'Dimensions': "Not available",
        'Classification': "Not available",
        'Credit Line': "Not available",
        'Object Number': "Not available"
    }

    if overview_section:
        # Extract <p> tags containing the artwork details
        detail_items = overview_section.find_all('p', class_='artwork-tombstone--item')
        
        for item in detail_items:
            # Extract the label and value from each <p>
            label = item.find('span', class_='artwork-tombstone--label')
            value = item.find('span', class_='artwork-tombstone--value')
            
            if label and value:
                # Add to the overview details dictionary
                key = label.text.strip().rstrip(":")
                overview_details[key] = value.text.strip()

    return {
        'Title': overview_details['Title'],
        'Artist': overview_details['Artist'],
        'Date': overview_details['Date'],
        'Medium': overview_details['Medium'],
        'Dimensions': overview_details['Dimensions'],
        'Classification': overview_details['Classification'],
        'Credit Line': overview_details['Credit Line'],
        'Object Number': overview_details['Object Number'],
        'Image URL': image_url,
        'Text Description': description
    }
```

```python
def save_to_csv(data, csv_file):
    # Prepare CSV header
    header = ['Title', 'Artist', 'Date', 'Medium', 'Dimensions', 'Classification', 'Credit Line', 'Object Number', 'Image URL', 'Text Description']

    # Open the CSV file for writing
    with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(header)  # Write the header

        # Write the data row
        writer.writerow([
            data['Title'],
            data['Artist'],
            data['Date'],
            data['Medium'],
            data['Dimensions'],
            data['Classification'],
            data['Credit Line'],
            data['Object Number'],
            data['Image URL'],
            data['Text Description']
        ])

# Example Usage
url = "https://www.metmuseum.org/art/collection/search/436508"
artwork_data = scrape_artwork_info(url)
save_to_csv(artwork_data, "artwork.csv")

print("Data saved to artwork.csv")
```
